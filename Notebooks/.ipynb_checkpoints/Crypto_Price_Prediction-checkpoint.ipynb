### Data Loading Pipeline

```python
# Data fetching functions
def find_first_valid_timestamp():
    """Find the earliest timestamp with valid Bitcoin price data from CryptoCompare."""
    # The ts number represents a UNIX timestamp, which is the number of seconds since January 1, 1970 (UTC).
    ts = 1388534400  # January 1, 2014
    while True:
      # ts + (3600 * 24) -> Adding one day worth's seconds to the ts. (Fetching data for 24 hour period)
        df = fetch_cryptocompare_data(ts, ts + 3600 * 24) # (start timestamp, end timestamp)
        if not df.empty and (df['close'] > 0).any(): # Check if atleast one closing price is present and df is not empty
            return df['time'].min() # Earliest time from that data
        ts += 3600 * 24
        time.sleep(0.1)

def fetch_cryptocompare_data(start_ts, end_ts):
    """Fetch hourly OHLCV data from CryptoCompare for BTC/USDT."""
    base_url = "https://min-api.cryptocompare.com/data/histohour"
    params = {"fsym": "BTC", "tsym": "USDT", "limit": 2000, "toTs": end_ts} #Tether (Stablecoin)
    data_list = []
    while True:
        response = requests.get(base_url, params=params)
        if response.status_code != 200:
            logger.error(f"CryptoCompare API error: {response.status_code}")
            break
        data = response.json()
        if data['Response'] != 'Success':
            logger.error(f"CryptoCompare response error: {data['Message']}")
            break
        df = pd.DataFrame(data['Data'])
        if df.empty:
            break
        data_list.append(df)
        earliest_ts = df['time'].min()
        if earliest_ts <= start_ts:
            break
        params['toTs'] = earliest_ts - 1
        time.sleep(0.1)
    if data_list:
        historical_df = pd.concat(data_list, ignore_index=True)
        historical_df = historical_df[historical_df['time'] >= start_ts]
        return historical_df
    return pd.DataFrame()

def fetch_binanceus_data(start_ts, end_ts):
    """Fetch hourly kline data from Binance.us for BTC/USDT."""
    base_url = "https://api.binance.us/api/v3/klines"
    params = {"symbol": "BTCUSDT", "interval": "1h", "limit": 1000, "startTime": start_ts * 1000, "endTime": end_ts * 1000}
    data_list = []
    while True:
        response = requests.get(base_url, params=params)
        if response.status_code != 200:
            logger.error(f"Binance.us API error: {response.status_code}")
            break
        data = response.json()
        if not data:
            break
        df = pd.DataFrame(data, columns=['open_time', 'open', 'high', 'low', 'close', 'volume', 'close_time', 'quote_asset_volume', 'number_of_trades', 'taker_buy_base', 'taker_buy_quote', 'ignore'])
        df['open_time'] = (df['open_time'] / 1000).astype(int)
        df = df[['open_time', 'open', 'high', 'low', 'close', 'quote_asset_volume']]
        data_list.append(df)
        last_ts = df['open_time'].max()
        if last_ts >= end_ts:
            break
        params['startTime'] = (last_ts + 1) * 1000
        time.sleep(0.1)
    if data_list:
        recent_df = pd.concat(data_list, ignore_index=True)
        recent_df = recent_df[recent_df['open_time'] <= end_ts]
        return recent_df
    return pd.DataFrame()

def fetch_and_combine_data():
    """Fetch and combine historical, current, and recent Bitcoin price data."""
    logger.info("Loading and combining datasets")
    current_df = pd.read_csv('/content/drive/MyDrive/CS554/candlesticks-h.csv')
    current_df = current_df.dropna(subset=['open', 'high', 'low', 'close'])
    current_df['timestamp'] = current_df['timestamp'].astype(int)

    min_ts = current_df['timestamp'].min()
    max_ts = current_df['timestamp'].max()

    historical_start = find_first_valid_timestamp()
    historical_end = min_ts - 3600
    logger.info(f"Fetching historical data from {historical_start} to {historical_end}")
    historical_df = fetch_cryptocompare_data(historical_start, historical_end)
    historical_df = historical_df.rename(columns={'time': 'timestamp', 'volumeto': 'volume'})
    historical_df = historical_df[['timestamp', 'open', 'high', 'low', 'close', 'volume']]
    for col in ['open', 'high', 'low', 'close', 'volume']:
        historical_df[col] = pd.to_numeric(historical_df[col], errors='coerce')
    historical_df = historical_df.dropna(subset=['close'])
    historical_df = historical_df[historical_df['close'] > 0]

    recent_start = max_ts + 3600
    recent_end = int(time.time())
    logger.info(f"Fetching recent data from {recent_start} to {recent_end}")
    recent_df = fetch_binanceus_data(recent_start, recent_end)
    recent_df = recent_df.rename(columns={'open_time': 'timestamp', 'quote_asset_volume': 'volume'})
    for col in ['open', 'high', 'low', 'close', 'volume']:
        recent_df[col] = pd.to_numeric(recent_df[col], errors='coerce')
    recent_df = recent_df.dropna(subset=['close'])
    recent_df = recent_df[recent_df['close'] > 0]

    for col in ['open', 'high', 'low', 'close', 'volume']:
        current_df[col] = pd.to_numeric(current_df[col], errors='coerce')

    combined_df = pd.concat([historical_df, current_df, recent_df], ignore_index=True)
    combined_df = combined_df.sort_values('timestamp').drop_duplicates(subset='timestamp', keep='first')
    combined_df.to_csv('/content/drive/MyDrive/CS554/final_data.csv', index=False)
    logger.info(f"Combined dataset saved with {len(combined_df)} rows")
    return combined_df
```

## PAST FEATURES (Historical Data for Modeling)

### `open`, `high`, `low`, `volume`
- **Definition**: Daily market open, highest, lowest prices, and traded volume.
- **Importance**: Core market data; these give essential info about price movement and investor activity.

---

### `ema_7d` (Exponential Moving Average - 7 days)
- **Definition**: Weighted average price over 7 days, with more weight on recent data.
- **Importance**: Detects short-term trends more responsively than simple moving averages.

---

### `rsi` (Relative Strength Index)
- **Definition**: Momentum oscillator (0–100), indicating overbought (>70) or oversold (<30) conditions.
- **Importance**: Helps anticipate reversals or trend continuations.

---

### `macd` / `macd_signal` (Moving Average Convergence Divergence)
- **Definition**: Trend-following momentum indicator showing the relationship between two EMAs (typically 12 and 26).
- **Importance**: Identifies trend changes and momentum strength.

---

### `bb_upper`, `bb_lower`, `bb_width` (Bollinger Bands)
- **Definition**: Bands around a moving average based on standard deviation. Width indicates volatility.
- **Importance**: Highlights overbought/oversold zones and market volatility.

---

### `volume_ema_7`
- **Definition**: EMA of trading volume over 7 days.
- **Importance**: Smoothed volume trends help detect buying/selling pressure shifts.

---

### `obv` (On-Balance Volume)
- **Definition**: Cumulative volume-based indicator to measure buying/selling pressure.
- **Importance**: Volume-based confirmation of price trends.

---

### `mfi` (Money Flow Index)
- **Definition**: RSI-like indicator using both price and volume.
- **Importance**: Useful for identifying potential reversals based on money flow.

---

### `stoch_k` / `stoch_d` (Stochastic Oscillator)
- **Definition**: Measures momentum by comparing closing price to recent highs/lows.
- **Importance**: Signals upward/downward momentum shifts.

---

### `cci` (Commodity Channel Index)
- **Definition**: Measures the price deviation from its average.
- **Importance**: Identifies overbought/oversold levels.

---

### `close_lag_1` to `close_lag_6`
- **Definition**: Past closing prices from 1 to 6 time steps ago.
- **Importance**: Essential for time-series modeling (e.g., autoregressive relationships).

---

### `ma_7` (Moving Average - 7 days)
- **Definition**: Simple average of the past 7 days’ closing prices.
- **Importance**: Smooths price action to reveal trends.

---

### `volatility`
- **Definition**: Standard deviation or other measure of price fluctuations.
- **Importance**: Captures market risk, which strongly affects trader behavior.

---

### `log_close`
- **Definition**: Natural logarithm of closing price.
- **Importance**: Normalizes prices for better modeling, especially with large value ranges.

---

### `log_returns`
- **Definition**: Log difference between closing prices.
- **Importance**: Stationary, more predictable than raw price changes.

---

### `realized_vol`
- **Definition**: Actual observed volatility over a certain period.
- **Importance**: Helps predict future volatility and risk.

---

## FUTURE FEATURES (Time-based Exogenous Variables)

### `hour`, `dayofweek`, `month`, `quarter`
- **Definition**: Temporal indicators.
- **Importance**: Capture cyclical or seasonality effects (e.g., weekday behavior vs. weekend, monthly patterns).

---

### `fourier_week`, `fourier_year`
- **Definition**: Fourier-transformed cyclical features for weekly/yearly patterns.
- **Importance**: Models seasonality in a smooth and continuous way, enhancing time-series forecasting.


```python
def preprocess_data(df, past_features, future_features):
    """Preprocess cryptocurrency data for time series forecasting."""
    logger.info("Preprocessing data")
    df['datetime'] = pd.to_datetime(df['timestamp'], unit='s')
    df = df[['datetime', 'open', 'close', 'high', 'low', 'volume']].sort_values('datetime')
    df = df.set_index('datetime').asfreq('h', method='ffill')

    if df.isna().any().any():
        logger.warning("NaNs found, filling with forward fill")
        df = df.ffill()

    # Stationarity check
    result = adfuller(df['close'].dropna())
    if result[1] > 0.05:
        logger.info("Series is non-stationary (p-value: %.4f), applying differencing" % result[1])
        ts_close = TimeSeries.from_dataframe(df.reset_index(), 'datetime', ['close'])
        ts_close_diff = ts_close.diff()
        df['close_diff'] = ts_close_diff.pd_series()
        joblib.dump({'differenced': True}, os.path.join(output_dir, 'stationarity_info.pkl'))
    else:
        logger.info("Series is stationary (p-value: %.4f)" % result[1])
        df['close_diff'] = df['close']
        joblib.dump({'differenced': False}, os.path.join(output_dir, 'stationarity_info.pkl'))

    # Technical indicators
    df['ema_7d'] = ta.ema(df['close'], length=168)
    df['rsi'] = ta.rsi(df['close'], length=14)
    macd = ta.macd(df['close'], fast=12, slow=26, signal=9)
    df['macd'] = macd['MACD_12_26_9']
    df['macd_signal'] = macd['MACDs_12_26_9']
    bb = ta.bbands(df['close'], length=20, std=2)
    df['bb_upper'] = bb['BBU_20_2.0']
    df['bb_lower'] = bb['BBL_20_2.0']
    df['bb_width'] = df['bb_upper'] - df['bb_lower']
    df['volume_ema_7'] = ta.ema(df['volume'], length=7)
    df['obv'] = ta.obv(df['close'], df['volume'])
    df['mfi'] = ta.mfi(high=df['high'], low=df['low'], close=df['close'], volume=df['volume'], length=14)
    stoch = ta.stoch(high=df['high'], low=df['low'], close=df['close'], k=14, d=3)
    df['stoch_k'] = stoch['STOCHk_14_3_3']
    df['stoch_d'] = stoch['STOCHd_14_3_3']
    df['cci'] = ta.cci(df['high'], df['low'], df['close'], length=20)

    for lag in range(1, 7):
        df[f'close_lag_{lag}'] = df['close'].shift(lag)

    df['ma_7'] = df['close'].rolling(window=7).mean()
    df['volatility'] = df['close'].rolling(window=7).std()
    df['log_close'] = np.log(df['close'])
    df['log_returns'] = df['log_close'].diff()
    df['realized_vol'] = df['log_returns'].rolling(window=168).std() * np.sqrt(168)

    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['month'] = df.index.month
    df['quarter'] = df.index.quarter
    df['fourier_week'] = df.index.hour / 24 + df.index.dayofweek / 7
    df['fourier_year'] = df.index.dayofyear / 365

    df = df.dropna().astype('float32')
    assert not df.isna().any().any() and not np.isinf(df).any().any(), "Data contains NaN or infinite values"
    assert all(col in df.columns for col in past_features + future_features), "Missing features after preprocessing"

    import seaborn as sns
    import matplotlib.pyplot as plt
    
    # Set professional style
    sns.set_style("whitegrid")
    sns.set_context("talk")
    
    # Refined color palette
    colors = {
        'rsi': '#1f77b4',             
        'overbought': '#d62728',      
        'oversold': '#2ca02c',        
        'macd': '#4B0082',            
        'macd_signal': '#FFA07A'     
    }
    
    # Create subplots
    fig, axs = plt.subplots(2, 1, figsize=(16, 10), sharex=True)
    
    # --- RSI Plot ---
    sns.lineplot(x=df.index, y='rsi', data=df, ax=axs[0], label='RSI',
                 color=colors['rsi'], linewidth=2, alpha=0.9)
    axs[0].axhline(70, color=colors['overbought'], linestyle='--', linewidth=1.2, label='Overbought (70)', alpha=0.7)
    axs[0].axhline(30, color=colors['oversold'], linestyle='--', linewidth=1.2, label='Oversold (30)', alpha=0.7)
    axs[0].set_title('📈 RSI Over Time', fontsize=18, weight='bold')
    axs[0].set_ylabel('RSI')
    axs[0].legend(loc='upper left')
    axs[0].grid(True)
    
    # --- MACD Plot ---
    # MACD - darker and bolder
    sns.lineplot(x=df.index, y='macd', data=df, ax=axs[1], label='MACD',
                 color=colors['macd'], linewidth=2.5, alpha=1.0)
    
    # MACD Signal - lighter and thinner
    sns.lineplot(x=df.index, y='macd_signal', data=df, ax=axs[1], label='MACD Signal',
                 color=colors['macd_signal'], linewidth=1.5, alpha=0.7)
    
    axs[1].set_title('📉 MACD & Signal Over Time', fontsize=18, weight='bold')
    axs[1].set_ylabel('MACD')
    axs[1].legend(loc='upper left')
    axs[1].grid(True)
    
    # Common x-axis label
    axs[1].set_xlabel('Time', fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


    return df


# Preprocess data
processed_df = preprocess_data(df, past_features, future_features)
```

# Tune and train 

```python
def objective(trial, model_class, train_series_scaled, val_series_scaled, train_past_cov_scaled, val_past_cov_scaled, train_future_cov_scaled=None, val_future_cov_scaled=None):
    """Objective function for Optuna hyperparameter tuning."""
    params = {
        'input_chunk_length': trial.suggest_categorical('input_chunk_length', [24, 48, 72, 96]),
        'num_blocks': trial.suggest_int('num_blocks', 2, 6) if model_class == NBEATSModel else None,
        'layer_widths': trial.suggest_categorical('layer_widths', [128, 256, 512, 1024]) if model_class == NBEATSModel else None,
        'hidden_size': trial.suggest_categorical('hidden_size', [32, 64, 128, 256]) if model_class == TFTModel else None,
        'lstm_layers': trial.suggest_int('lstm_layers', 1, 4) if model_class == TFTModel else None,
        'batch_size': trial.suggest_categorical('batch_size', [16, 32, 64]),
    }
    params = {k: v for k, v in params.items() if v is not None}

    # Use a distinct name for the CSVLogger instance
    trial_logger = CSVLogger(save_dir=output_dir, name=f"{model_class.__name__}_logs", version=trial.number)
    callbacks = [
        EarlyStopping(monitor="val_loss", patience=5, mode="min"),
        ModelCheckpoint(dirpath=output_dir, filename=f"{model_class.__name__}_trial_{trial.number}", monitor="val_loss", mode="min")
    ]

    model = model_class(
        output_chunk_length=24,
        likelihood=QuantileRegression(quantiles=[0.05, 0.5, 0.95]),
        **params,
        n_epochs=20,
        random_state=seed,
        pl_trainer_kwargs={"accelerator": "gpu", "callbacks": callbacks, "logger": trial_logger, "enable_checkpointing": True}
    )

    try:
        if model_class == NBEATSModel:
            model.fit(series=train_series_scaled, past_covariates=train_past_cov_scaled, val_series=val_series_scaled, val_past_covariates=val_past_cov_scaled)
        else:
            model.fit(series=train_series_scaled, past_covariates=train_past_cov_scaled, future_covariates=train_future_cov_scaled,
                      val_series=val_series_scaled, val_past_covariates=val_past_cov_scaled, val_future_covariates=val_future_cov_scaled)
    except Exception as e:
        # Use the global logger for error logging
        logger.error(f"Training failed: {e}")
        raise

    forecasts_scaled = model.historical_forecasts(
        series=val_series_scaled, past_covariates=val_past_cov_scaled, future_covariates=val_future_cov_scaled if model_class == TFTModel else None,
        forecast_horizon=1, stride=1, retrain=False
    )
    forecasts_original = series_scaler.inverse_transform(forecasts_scaled)
    mae_value = mae(val_series, forecasts_original)
    # Use the global logger for info logging
    logger.info(f"Trial {trial.number} params: {params}, MAE: {mae_value}")
    return mae_value
```

```python
import logging
import torch
from pytorch_lightning.callbacks import EarlyStopping, ModelCheckpoint
from pytorch_lightning.loggers import CSVLogger
import pandas as pd
import matplotlib.pyplot as plt
import os
from darts.models import NBEATSModel, TFTModel
from darts.utils.likelihood_models import QuantileRegression

# --- Logging Setup ---
logger = logging.getLogger(__name__)
logger.setLevel(logging.INFO)

formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")

# Console handler (prints logs)
console_handler = logging.StreamHandler()
console_handler.setFormatter(formatter)
logger.addHandler(console_handler)

# File handlers (saves logs to both .log and .txt)
log_file_handler = logging.FileHandler("training.log")
log_file_handler.setFormatter(formatter)
logger.addHandler(log_file_handler)

txt_file_handler = logging.FileHandler("training.txt")
txt_file_handler.setFormatter(formatter)
logger.addHandler(txt_file_handler)

# --- Main Function ---
def train_model(model_class, train_series_scaled, val_series_scaled, train_past_cov_scaled, val_past_cov_scaled, train_future_cov_scaled=None, val_future_cov_scaled=None, output_dir="models/", seed=42):
    """
    Train an N-BEATS or TFT model with predefined hyperparameters.

    Parameters:
    - model_class: The model class to train (NBEATSModel or TFTModel).
    - train_series_scaled: Scaled training time series (Darts TimeSeries object).
    - val_series_scaled: Scaled validation time series (Darts TimeSeries object).
    - train_past_cov_scaled: Scaled past covariates for training (Darts TimeSeries object).
    - val_past_cov_scaled: Scaled past covariates for validation (Darts TimeSeries object).
    - train_future_cov_scaled: Scaled future covariates for training (required for TFTModel, optional).
    - val_future_cov_scaled: Scaled future covariates for validation (required for TFTModel, optional).
    - output_dir: Directory to save model and logs (default: "models/").
    - seed: Random seed for reproducibility (default: 42).

    Returns:
    - Trained model instance.
    """
    
    # Define default parameters based on model class
    if model_class == NBEATSModel:
        default_params = {
            'input_chunk_length': 24,  # Length of input sequences
            'num_blocks': 3,          # Number of blocks in N-BEATS architecture
            'layer_widths': 256,      # Width of hidden layers
            'batch_size': 32          # Batch size for training
        }
    elif model_class == TFTModel:
        default_params = {
            'input_chunk_length': 24,  # Length of input sequences
            'hidden_size': 64,        # Size of hidden layers
            'lstm_layers': 2,         # Number of LSTM layers
            'batch_size': 32          # Batch size for training
        }
    else:
        raise ValueError("Unsupported model class. Use NBEATSModel or TFTModel.")

    # Set up logger and callbacks for training
    logger_final = CSVLogger(save_dir=output_dir, name=f"{model_class.__name__}_final")
    callbacks = [
        EarlyStopping(monitor="val_loss", patience=10, mode="min"),
        ModelCheckpoint(dirpath=output_dir, filename=f"{model_class.__name__}_best", monitor="val_loss", mode="min")
    ]

    # Log device availability
    if torch.backends.mps.is_available():
        logger.info("MPS backend is available, attempting to use GPU acceleration.")
    else:
        logger.info("MPS backend is not available, falling back to CPU.")

    # Initialize the model with predefined parameters
    model = model_class(
        output_chunk_length=24,  # Length of forecast horizon
        likelihood=QuantileRegression(quantiles=[0.05, 0.5, 0.95]),  # Quantile regression for probabilistic forecasts
        **default_params,
        n_epochs=20,  # Number of training epochs
        random_state=seed,
        pl_trainer_kwargs={
            "accelerator": "cpu",        
            "callbacks": callbacks,       # Early stopping and checkpointing
            "logger": logger_final,       # Log metrics to CSV
            "enable_checkpointing": True  # Enable model checkpointing
        }
    )

    # Train the model
    try:
        if model_class == NBEATSModel:
            model.fit(
                series=train_series_scaled,
                past_covariates=train_past_cov_scaled,
                val_series=val_series_scaled,
                val_past_covariates=val_past_cov_scaled
            )
        else:  # TFTModel
            if train_future_cov_scaled is None or val_future_cov_scaled is None:
                raise ValueError("TFTModel requires future covariates.")
            model.fit(
                series=train_series_scaled,
                past_covariates=train_past_cov_scaled,
                future_covariates=train_future_cov_scaled,
                val_series=val_series_scaled,
                val_past_covariates=val_past_cov_scaled,
                val_future_covariates=val_future_cov_scaled
            )
        logger.info(f"Training completed for {model_class.__name__}.")
    except Exception as e:
        logger.error(f"Training failed for {model_class.__name__}: {e}")
        raise

    # Plot training and validation loss
    metrics_path = f"{output_dir}/{model_class.__name__}_final/version_0/metrics.csv"
    if os.path.exists(metrics_path):
        metrics = pd.read_csv(metrics_path)
        metrics = metrics.groupby('epoch').mean().reset_index()
        plt.figure(figsize=(10, 5))
        plt.plot(metrics['epoch'], metrics['train_loss'], label='Train Loss')
        plt.plot(metrics['epoch'], metrics['val_loss'], label='Val Loss')
        plt.legend()
        plt.title(f'{model_class.__name__} Training and Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.show()
    else:
        logger.warning(f"Metrics file not found: {metrics_path}")

    # Save the trained model
    save_path = os.path.join(output_dir, f"{model_class.__name__}_model.pkl")
    model.save(save_path)
    logger.info(f"Model saved to {save_path}")

    return model

# nbeats_model = train_model(NBEATSModel, train_series_scaled, val_series_scaled, train_past_cov_scaled, val_past_cov_scaled)
# tft_model = train_model(TFTModel, train_series_scaled, val_series_scaled, train_past_cov_scaled, val_past_cov_scaled, train_future_cov_scaled, val_future_cov_scaled)
```

```python
import numpy as np
import pandas as pd
import logging
import torch

# Set up logging
logger = logging.getLogger("DebugLogger")
logger.setLevel(logging.INFO)
ch = logging.StreamHandler()
ch.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))
logger.addHandler(ch)

def check_time_series_stats(ts, name="TimeSeries"):
    """
    Log summary statistics for a Darts TimeSeries object.
    If the series is multivariate, it converts it to a DataFrame and prints stats per column.
    """
    try:
        # Check if the TimeSeries is univariate
        if ts.width == 1:
            # For univariate, we can convert to a pandas Series using to_series()
            series = ts.to_series()
            logger.info(f"{name} - min: {series.min()}, max: {series.max()}, mean: {series.mean()}, std: {series.std()}")
            if series.isna().sum() > 0:
                logger.warning(f"{name} contains {series.isna().sum()} NaN values.")
            if np.isinf(series).sum() > 0:
                logger.warning(f"{name} contains infinite values.")
        else:
            # For multivariate series, convert to a DataFrame
            df = ts.pd_dataframe()
            logger.info(f"{name} summary statistics:")
            logger.info(df.describe())
            if df.isna().any().any():
                logger.warning(f"{name} contains NaN values.")
            if np.isinf(df.values).any():
                logger.warning(f"{name} contains infinite values.")
    except Exception as e:
        logger.error(f"Error checking stats for {name}: {e}")

def check_alignment(series, past_cov, future_cov=None, name="Alignment Check"):
    """Ensure that the time indices of series and covariates are aligned."""
    series_idx = series.time_index
    past_idx = past_cov.time_index
    logger.info(f"{name} - Series: {series_idx[0]} to {series_idx[-1]}")
    logger.info(f"{name} - Past covariates: {past_idx[0]} to {past_idx[-1]}")
    
    if not series_idx.equals(past_idx):
        logger.warning("Time index mismatch between series and past covariates.")
    
    if future_cov is not None:
        future_idx = future_cov.time_index
        logger.info(f"{name} - Future covariates: {future_idx[0]} to {future_idx[-1]}")
        if not series_idx.equals(future_idx):
            logger.warning("Time index mismatch between series and future covariates.")

# Example usage for your scaled TimeSeries objects:
logger.info("=== Checking Scaled TimeSeries Statistics ===")
check_time_series_stats(train_series_scaled, "Train Series")
check_time_series_stats(val_series_scaled, "Validation Series")
check_time_series_stats(train_past_cov_scaled, "Train Past Covariates")
check_time_series_stats(val_past_cov_scaled, "Validation Past Covariates")
if 'train_future_cov_scaled' in globals() and train_future_cov_scaled is not None:
    check_time_series_stats(train_future_cov_scaled, "Train Future Covariates")
if 'val_future_cov_scaled' in globals() and val_future_cov_scaled is not None:
    check_time_series_stats(val_future_cov_scaled, "Validation Future Covariates")

logger.info("=== Checking Alignment of Time Indices ===")
check_alignment(train_series_scaled, train_past_cov_scaled, train_future_cov_scaled)
check_alignment(val_series_scaled, val_past_cov_scaled, val_future_cov_scaled)
```

```python
def plot_forecasts_with_intervals(model, val_series_scaled, val_series, close_series, val_past_cov_scaled, val_future_cov_scaled=None):
    # Generate historical forecasts
    forecasts = model.historical_forecasts(
        series=val_series_scaled,
        past_covariates=val_past_cov_scaled,
        future_covariates=val_future_cov_scaled if model.supports_future_covariates else None,
        forecast_horizon=1,
        stride=1,
        retrain=False
    )
    
    # Inverse transform to original scale
    forecasts_original = series_scaler.inverse_transform(forecasts)
    
    # Log forecast shape for debugging
    logger.info(f"forecasts_original.values().shape: {forecasts_original.values().shape}")
    logger.info(f"forecasts_original.components: {forecasts_original.components}")
    
    # Extract forecast values based on dimensionality
    forecast_values = forecasts_original.values()
    time_index = forecasts_original.time_index
    
    if forecast_values.ndim == 3 and forecast_values.shape[2] == 3:
        # Probabilistic forecasts with quantiles
        lower_bound_diff_values = forecast_values[:, 0, 0]  # 0.05 quantile
        median_forecast_diff_values = forecast_values[:, 0, 1]  # 0.5 quantile
        upper_bound_diff_values = forecast_values[:, 0, 2]  # 0.95 quantile
    elif forecast_values.ndim == 2 and forecast_values.shape[1] == 1:
        # Deterministic forecast
        logger.warning("Probabilistic forecasts not available; plotting median forecast only.")
        median_forecast_diff_values = forecast_values[:, 0]
        lower_bound_diff_values = median_forecast_diff_values
        upper_bound_diff_values = median_forecast_diff_values
    else:
        raise ValueError("Unexpected forecast shape. Expected 2D (deterministic) or 3D (probabilistic) array.")

    # Create TimeSeries objects
    lower_bound_diff = TimeSeries.from_times_and_values(time_index, lower_bound_diff_values)
    median_forecast_diff = TimeSeries.from_times_and_values(time_index, median_forecast_diff_values)
    upper_bound_diff = TimeSeries.from_times_and_values(time_index, upper_bound_diff_values)
    
    # Convert differences back to close prices
    lower_bound_close = []
    median_forecast_close = []
    upper_bound_close = []
    for t in time_index:
        prev_time = t - pd.Timedelta(hours=1)
        if prev_time in close_series.time_index:
            prev_close = close_series[prev_time].values()[0]
            lower_bound_close.append(prev_close + lower_bound_diff[t].values()[0])
            median_forecast_close.append(prev_close + median_forecast_diff[t].values()[0])
            upper_bound_close.append(prev_close + upper_bound_diff[t].values()[0])
    
    # Create TimeSeries for close prices
    lower_bound_close_ts = TimeSeries.from_times_and_values(time_index, lower_bound_close)
    median_forecast_close_ts = TimeSeries.from_times_and_values(time_index, median_forecast_close)
    upper_bound_close_ts = TimeSeries.from_times_and_values(time_index, upper_bound_close)
    
    # Plotting
    plt.figure(figsize=(12, 6))
    close_series[time_index].plot(label='Actual Close')
    median_forecast_close_ts.plot(label='Median Forecast Close')
    if forecast_values.ndim == 3 and forecast_values.shape[2] == 3:
        plt.fill_between(
            time_index,
            lower_bound_close_ts.values().flatten(),
            upper_bound_close_ts.values().flatten(),
            alpha=0.3,
            label='90% Prediction Interval'
        )
    plt.legend()
    plt.title(f'{model.__class__.__name__} Forecasts for Close Price')
    plt.xlabel('Time')
    plt.ylabel('Close Price (USD)')
    plt.show()


def plot_forecasts_with_intervals_and_actuals_separately(
    model,
    val_series_scaled,
    val_series,
    close_series,
    val_past_cov_scaled,
    val_future_cov_scaled=None,
    series_scaler=None
):
    import matplotlib.pyplot as plt
    from darts import TimeSeries

    # --- Forecast Generation ---
    forecasts = model.historical_forecasts(
        series=val_series_scaled,
        past_covariates=val_past_cov_scaled,
        future_covariates=val_future_cov_scaled if model.supports_future_covariates else None,
        forecast_horizon=1,
        stride=1,
        retrain=False
    )

    forecasts_original = series_scaler.inverse_transform(forecasts)
    forecast_values = forecasts_original.values()
    time_index = forecasts_original.time_index

    # --- Extract Forecasts ---
    if forecast_values.ndim == 3 and forecast_values.shape[2] == 3:
        lower = forecast_values[:, 0, 0]
        median = forecast_values[:, 0, 1]
        upper = forecast_values[:, 0, 2]
    else:
        logger.warning("Model is deterministic; no prediction intervals.")
        median = forecast_values[:, 0]
        lower = upper = median

    # --- Convert diff → price ---
    lower_close, median_close, upper_close = [], [], []
    for t in time_index:
        prev_time = t - pd.Timedelta(hours=1)
        if prev_time in close_series.time_index:
            prev_close = close_series[prev_time].values()[0]
            idx = time_index.get_loc(t)
            lower_close.append(prev_close + lower[idx])
            median_close.append(prev_close + median[idx])
            upper_close.append(prev_close + upper[idx])

    lower_ts = TimeSeries.from_times_and_values(time_index, lower_close)
    median_ts = TimeSeries.from_times_and_values(time_index, median_close)
    upper_ts = TimeSeries.from_times_and_values(time_index, upper_close)

    # --- Plot 1: Forecast Only ---
    plt.figure(figsize=(12, 6))
    median_ts.plot(label="Median Forecast")
    if forecast_values.ndim == 3:
        plt.fill_between(
            time_index,
            lower_ts.values().flatten(),
            upper_ts.values().flatten(),
            alpha=0.3,
            color="orange",
            label="90% Prediction Interval"
        )
    plt.title(f"{model.__class__.__name__} Forecast (Median + Interval)")
    plt.xlabel("Time")
    plt.ylabel("Predicted Close Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # --- Plot 2: Actual Close Price Only ---
    plt.figure(figsize=(12, 5))
    close_series[time_index].plot(label="Actual Close Price")
    plt.title("📈 Actual Bitcoin Close Price")
    plt.xlabel("Time")
    plt.ylabel("Close Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def evaluate_and_plot_models(
    val_series_scaled: TimeSeries,
    close_series: TimeSeries,
    val_past_cov_scaled: TimeSeries,
    val_future_cov_scaled: TimeSeries,
    series_scaler,
    nbeats_model: NBEATSModel,
    tft_model: TFTModel
) -> Tuple[Dict, Dict, Dict]:
    def get_forecast(model, model_type: str) -> TimeSeries:
        try:
            forecast_scaled = model.historical_forecasts(
                series=val_series_scaled,
                past_covariates=val_past_cov_scaled if model.supports_past_covariates else None,
                future_covariates=val_future_cov_scaled if model.supports_future_covariates else None,
                forecast_horizon=1,
                stride=1,
                retrain=False,
                verbose=True
            )
            forecast = series_scaler.inverse_transform(forecast_scaled)
            return forecast.quantile_timeseries(0.5) if forecast.is_probabilistic else forecast
        except Exception as e:
            logger.error(f"Error generating {model_type} forecast: {str(e)}")
            raise

    def invert_diff(forecast_diff: TimeSeries, original_series: TimeSeries) -> TimeSeries:
        try:
            forecast_times = forecast_diff.time_index
            prev_times = forecast_times - pd.Timedelta(hours=1)
            prev_values = [
                original_series[pt].values()[0] if pt in original_series.time_index else np.nan
                for pt in prev_times
            ]
            aligned_prev = TimeSeries.from_times_and_values(forecast_times, prev_values)
            combined = aligned_prev + forecast_diff
            # Drop missing values manually
            valid_mask = ~np.isnan(combined.univariate_values())
            return TimeSeries.from_times_and_values(combined.time_index[valid_mask], combined.univariate_values()[valid_mask])
        except Exception as e:
            logger.error(f"Error inverting differenced forecast: {str(e)}")
            raise

    logger.info("Generating model forecasts...")
    nbeats_forecast_diff = get_forecast(nbeats_model, "N-BEATS")
    tft_forecast_diff = get_forecast(tft_model, "TFT")

    start_date = max(nbeats_forecast_diff.start_time(), tft_forecast_diff.start_time())
    end_date = min(nbeats_forecast_diff.end_time(), tft_forecast_diff.end_time())

    nbeats_aligned = nbeats_forecast_diff.slice(start_date, end_date)
    tft_aligned = tft_forecast_diff.slice(start_date, end_date)
    combined_forecast_diff = (nbeats_aligned + tft_aligned) * 0.5

    actual_close = close_series.slice(start_date, end_date)

    nbeats_forecast = invert_diff(nbeats_aligned, close_series)
    tft_forecast = invert_diff(tft_aligned, close_series)
    combined_forecast = invert_diff(combined_forecast_diff, close_series)

    def strict_align(*series: TimeSeries) -> List[TimeSeries]:
        common_start = max(s.start_time() for s in series)
        common_end = min(s.end_time() for s in series)
        return [s.slice(common_start, common_end) for s in series]

    actual_close, nbeats_forecast, tft_forecast, combined_forecast = strict_align(
        actual_close, nbeats_forecast, tft_forecast, combined_forecast
    )

    def smape(actual, forecast):
        actual_vals = actual.univariate_values()
        forecast_vals = forecast.univariate_values()
        return 100.0 * np.mean(
            2.0 * np.abs(actual_vals - forecast_vals) / (np.abs(actual_vals) + np.abs(forecast_vals))
        )

    def mse(actual, forecast):
        return np.mean((actual.univariate_values() - forecast.univariate_values()) ** 2)

    def compute_metrics(actual: TimeSeries, forecast: TimeSeries) -> Dict[str, float]:
        actual_vals = actual.univariate_values()
        forecast_vals = forecast.univariate_values()
        mae_val = mae(actual, forecast)
        rmse_val = rmse(actual, forecast)
        mse_val = mse(actual, forecast)
        mape_val = mape(actual, forecast)
        smape_val = smape(actual, forecast)
        r2_val = r2_score(actual, forecast)
        return {
            'MAE': mae_val,
            'RMSE': rmse_val,
            'MSE': mse_val,
            'MAPE': mape_val,
            'SMAPE': smape_val,
            'R2': r2_val
        }

    metrics_nbeats = compute_metrics(actual_close, nbeats_forecast)
    metrics_tft = compute_metrics(actual_close, tft_forecast)
    metrics_combined = compute_metrics(actual_close, combined_forecast)

    def safe_subtract(a: TimeSeries, b: TimeSeries) -> TimeSeries:
        a, b = a.slice_intersect(b), b.slice_intersect(a)
        return a - b

    plt.figure(figsize=(16, 9))
    actual_close.plot(label='Actual Price', color='black', linewidth=2)
    nbeats_forecast.plot(label='N-BEATS Forecast', color='blue', linewidth=1.5, alpha=0.8)
    tft_forecast.plot(label='TFT Forecast', color='green', linewidth=1.5, alpha=0.8)
    combined_forecast.plot(label='Combined Forecast', color='red', linestyle='--', linewidth=1.8, alpha=0.7)
    plt.title(f'Price Forecast Comparison\n{start_date} to {end_date}', fontsize=14)
    plt.legend(loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.show()

    fig, axs = plt.subplots(3, 1, figsize=(14, 12), sharex=True)
    for ax, (name, residuals) in zip(axs, [
        ('N-BEATS', safe_subtract(actual_close, nbeats_forecast)),
        ('TFT', safe_subtract(actual_close, tft_forecast)),
        ('Combined', safe_subtract(actual_close, combined_forecast))
    ]):
        residuals.plot(ax=ax, linewidth=1.2)
        ax.set_title(f'{name} Residuals', fontsize=12)
        ax.set_ylabel('Prediction Error')
        ax.grid(True, linestyle='--', alpha=0.4)
    axs[-1].set_xlabel('Time')
    plt.tight_layout()
    plt.show()

    print("\n=== N-BEATS Metrics ===")
    for k, v in metrics_nbeats.items():
        print(f"{k}: {v:.4f}")

    print("\n=== TFT Metrics ===")
    for k, v in metrics_tft.items():
        print(f"{k}: {v:.4f}")

    print("\n=== Combined Model Metrics ===")
    for k, v in metrics_combined.items():
        print(f"{k}: {v:.4f}")

    return metrics_nbeats, metrics_tft, metrics_combined
```

```python
def preprocess_new_data(df, past_features, future_features, input_chunk_length):
    """Preprocess new incoming data."""
    processed_data = preprocess_data(df, past_features, future_features)
    return processed_data.tail(input_chunk_length)

def predict_next_hour_close(
    processed_data,  # <-- Now this is passed in
    model_name='combined',
    input_chunk_length=28
):
    """Predict the next hour's closing price using preprocessed data.

    Args:
        processed_data: Preprocessed DataFrame (with required columns and tail trimmed).
        model_name: Model to use for prediction ('nbeats', 'tft', or 'combined'). Default is 'combined'.
        input_chunk_length: Number of past time steps to use as input.

    Returns:
        float: Predicted closing price for the next hour.
    """
    global nbeats_model, tft_model, output_dir, past_features, future_features

    # Check for NaN values in processed data
    if processed_data.isnull().values.any():
        logger.error("Processed data contains NaN values. Check raw data or preprocessing steps.")
        raise ValueError("Processed data contains NaN values")

    # Convert to TimeSeries objects
    series = TimeSeries.from_dataframe(
        processed_data.reset_index(), 'datetime', ['close_diff'], freq='h'
    ).astype(np.float32)
    past_cov = TimeSeries.from_dataframe(
        processed_data.reset_index(), 'datetime', past_features, freq='h'
    ).astype(np.float32)

    # Load scalers
    try:
        series_scaler = joblib.load(os.path.join(output_dir, 'series_scaler.pkl'))
        past_cov_scaler = joblib.load(os.path.join(output_dir, 'past_cov_scaler.pkl'))
        if model_name in ['tft', 'combined']:
            future_cov_scaler = joblib.load(os.path.join(output_dir, 'future_cov_scaler.pkl'))
    except FileNotFoundError as e:
        logger.error(f"Scaler file not found: {e}")
        raise

    # Scale the data
    series_scaled = series_scaler.transform(series)
    past_cov_scaled = past_cov_scaler.transform(past_cov)

    if np.isnan(series_scaled.values()).any():
        logger.warning("Scaled series contains NaN values")
    if np.isnan(past_cov_scaled.values()).any():
        logger.warning("Scaled past covariates contain NaN")

    # Prediction logic
    if model_name == 'nbeats':
        pred_scaled = nbeats_model.predict(
            n=1, series=series_scaled, past_covariates=past_cov_scaled
        )
        pred = series_scaler.inverse_transform(pred_scaled)

    elif model_name == 'tft':
        output_chunk_length = tft_model.output_chunk_length
        start_time = processed_data.index[0]
        end_time = processed_data.index[-1] + pd.Timedelta(hours=output_chunk_length)
        future_times = pd.date_range(start=start_time, end=end_time, freq='h')
        future_df = pd.DataFrame({
            'datetime': future_times,
            'hour': future_times.hour,
            'dayofweek': future_times.dayofweek,
            'month': future_times.month,
            'quarter': future_times.quarter,
            'fourier_week': future_times.hour / 24 + future_times.dayofweek / 7,
            'fourier_year': future_times.dayofyear / 365
        })
        future_cov = TimeSeries.from_dataframe(
            future_df, 'datetime', future_features, freq='h'
        ).astype(np.float32)
        future_cov_scaled = future_cov_scaler.transform(future_cov)

        pred_scaled = tft_model.predict(
            n=1, series=series_scaled, past_covariates=past_cov_scaled, future_covariates=future_cov_scaled
        )
        pred = series_scaler.inverse_transform(pred_scaled)

    elif model_name == 'combined':
        pred_nbeats_scaled = nbeats_model.predict(
            n=1, series=series_scaled, past_covariates=past_cov_scaled
        )
        pred_nbeats = series_scaler.inverse_transform(pred_nbeats_scaled)

        output_chunk_length = tft_model.output_chunk_length
        start_time = processed_data.index[0]
        end_time = processed_data.index[-1] + pd.Timedelta(hours=output_chunk_length)
        future_times = pd.date_range(start=start_time, end=end_time, freq='h')
        future_df = pd.DataFrame({
            'datetime': future_times,
            'hour': future_times.hour,
            'dayofweek': future_times.dayofweek,
            'month': future_times.month,
            'quarter': future_times.quarter,
            'fourier_week': future_times.hour / 24 + future_times.dayofweek / 7,
            'fourier_year': future_times.dayofyear / 365
        })
        future_cov = TimeSeries.from_dataframe(
            future_df, 'datetime', future_features, freq='h'
        ).astype(np.float32)
        future_cov_scaled = future_cov_scaler.transform(future_cov)

        pred_tft_scaled = tft_model.predict(
            n=1, series=series_scaled, past_covariates=past_cov_scaled, future_covariates=future_cov_scaled
        )
        pred_tft = series_scaler.inverse_transform(pred_tft_scaled)

        # Combine predictions
        pred = pred_nbeats * 0.5 + pred_tft * 0.5

    else:
        raise ValueError("model_name must be 'nbeats', 'tft', or 'combined'")

    diff_pred = pred.values()[0][0].item()
    last_close = processed_data['close'].iloc[-1]
    stationarity_info = joblib.load(os.path.join(output_dir, 'stationarity_info.pkl'))

    actual_pred = last_close + diff_pred if stationarity_info['differenced'] else diff_pred
    logger.info(f"Predicted next hour's price: {actual_pred:.2f} USD")
    return actual_pred




new_data = df.tail(201).copy()
processed = preprocess_new_data(new_data, past_features, future_features, input_chunk_length=28)

try:
    pred = predict_next_hour_close(processed, model_name='combined')
    print(f"Predicted next hour's price: {pred:.2f} USD")
except Exception as e:
    logger.error(f"Prediction failed: {e}")
```

### Evaluate on Recent Data 

```python
def evaluate_on_recent_data(n_hours: int = 168):
    """
    Fetch and evaluate model performance on recent, unseen BTC/USDT data.

    Args:
        n_hours (int): How many hours of recent data to evaluate on

    Returns:
        Tuple[Dict, Dict, Dict]: metrics for N-BEATS, TFT, Combined
    """
    import time
    import pandas as pd
    from darts import TimeSeries

    # Get timestamps after training set
    last_train_time = train_series.end_time()
    start_ts = int((last_train_time + pd.Timedelta(hours=1)).timestamp())
    end_ts = int(time.time())

    logger.info(f"Fetching recent data from {start_ts} to {end_ts}")
    raw_df = fetch_binanceus_data(start_ts, end_ts)

    if raw_df.empty:
        raise ValueError("No recent data returned from Binance.")

    raw_df = raw_df.rename(columns={'open_time': 'timestamp', 'quote_asset_volume': 'volume'})
    for col in ['open', 'high', 'low', 'close', 'volume']:
        raw_df[col] = pd.to_numeric(raw_df[col], errors='coerce')
    raw_df = raw_df.dropna(subset=['close'])
    raw_df = raw_df[raw_df['close'] > 0]

    # Preprocess
    processed_df = preprocess_data(raw_df, past_features, future_features)

    # Use only the last `n_hours` for evaluation
    processed_df = processed_df.tail(n_hours)

    if processed_df.empty:
        raise ValueError("Processed recent data is empty after trimming.")

    # TimeSeries conversion
    stationarity_info = joblib.load(os.path.join(output_dir, 'stationarity_info.pkl'))
    differenced = stationarity_info.get('differenced', False)
    close_col = 'close_diff' if differenced else 'close'

    recent_series = TimeSeries.from_dataframe(processed_df.reset_index(), 'datetime', [close_col], freq='h')
    recent_close = TimeSeries.from_dataframe(processed_df.reset_index(), 'datetime', ['close'], freq='h')
    recent_past_cov = TimeSeries.from_dataframe(processed_df.reset_index(), 'datetime', past_features, freq='h')
    recent_future_cov = TimeSeries.from_dataframe(processed_df.reset_index(), 'datetime', future_features, freq='h')

    # Clip values to training bounds
    min_val, max_val = train_series.min().values()[0][0], train_series.max().values()[0][0]
    clipped = recent_series.pd_dataframe().clip(lower=min_val, upper=max_val)
    recent_series_clipped = TimeSeries.from_dataframe(clipped.reset_index(), 'datetime', value_cols=[close_col], freq='h')

    # Scale using saved scalers
    recent_series_scaled = series_scaler.transform(recent_series_clipped)
    recent_past_cov_scaled = past_cov_scaler.transform(recent_past_cov)
    recent_future_cov_scaled = future_cov_scaler.transform(recent_future_cov)

    # Evaluate using original evaluation pipeline
    logger.info("Evaluating models on recent unseen data...")
    return evaluate_and_plot_models(
        val_series_scaled=recent_series_scaled,
        close_series=recent_close,
        val_past_cov_scaled=recent_past_cov_scaled,
        val_future_cov_scaled=recent_future_cov_scaled,
        series_scaler=series_scaler,
        nbeats_model=nbeats_model,
        tft_model=tft_model
    )
```
