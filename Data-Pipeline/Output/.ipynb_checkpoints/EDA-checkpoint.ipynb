```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# --- CONFIG ---
csv_path = "data-merged.csv" 

# --- Load & Clean Data ---
def load_and_clean_data(csv_path):
    df = pd.read_csv(csv_path)

    # --- Parse datetime ---
    try:
        df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    except:
        pass

    if pd.api.types.is_numeric_dtype(df['timestamp']):
        if df['timestamp'].max() > 1e12:
            df['datetime'] = pd.to_datetime(df['timestamp'], unit='ms')
        else:
            df['datetime'] = pd.to_datetime(df['timestamp'], unit='s')
    else:
        df['datetime'] = pd.to_datetime(df['timestamp'], errors='coerce')

    # --- Drop duplicate timestamps ---
    df = df.drop_duplicates(subset='datetime')
    df = df.set_index('datetime').sort_index()
    df = df.asfreq('h')  # Ensure consistent hourly frequency

    # --- Drop constant columns and fill missing values ---
    constant_cols = [col for col in df.columns if df[col].std() == 0]
    df.drop(columns=constant_cols, inplace=True)
    df = df.replace([np.inf, -np.inf], np.nan).fillna(0)

    return df

# --- Visualization Functions ---
def plot_feature_trends(df, columns):
    df[columns].plot(subplots=True, figsize=(12, 8), title="Time Series Trends")
    plt.tight_layout()
    plt.show()

def plot_scatter_features(df, target, features):
    for col in features:
        plt.figure()
        sns.scatterplot(x=df[col], y=df[target])
        plt.title(f"{col} vs {target}")
        plt.grid(True)
        plt.tight_layout()
        plt.show()

def plot_acf_pacf(series, lags=48):
    plot_acf(series.dropna(), lags=lags)
    plt.title("Autocorrelation (ACF)")
    plt.show()

    plot_pacf(series.dropna(), lags=lags)
    plt.title("Partial Autocorrelation (PACF)")
    plt.show()

# --- Main Analysis ---
def analyze_patterns(csv_path):
    df = load_and_clean_data(csv_path)

    #  Trend Plots
    trend_cols = ['close', 'volume', 'median_sentiment', 'rsi', 'macd', 'bb_width']
    available_trend_cols = [col for col in trend_cols if col in df.columns]
    if available_trend_cols:
        plot_feature_trends(df, available_trend_cols)

    # ACF/PACF of close_diff
    if 'close_diff' in df.columns:
        plot_acf_pacf(df['close_diff'])

    #  Scatter plots of features vs close_diff
    scatter_features = ['volume', 'rsi', 'macd', 'median_sentiment', 'bb_width']
    available_features = [f for f in scatter_features if f in df.columns]
    if 'close_diff' in df.columns and available_features:
        plot_scatter_features(df, target='close_diff', features=available_features)

analyze_patterns(csv_path)
```
